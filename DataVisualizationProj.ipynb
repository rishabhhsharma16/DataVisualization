## Python Code to Visualize DataSets

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings as w
w.filterwarnings('ignore')
```

```python
aq1=pd.read_csv(r'C:\Users\Default.DESKTOP-A7J9TOP\Desktop\DS Project\BSEINDEX.csv',parse_dates=['Month'],index_col='Month')
aq3=pd.read_csv(r"C:\Users\Default.DESKTOP-A7J9TOP\Desktop\DS Project\CRUDEOILDOLLAR2.csv",parse_dates=['Month'],index_col='Month')
aq2=pd.read_csv(r'C:\Users\Default.DESKTOP-A7J9TOP\Desktop\DS Project\GOLDPRICE2.csv',parse_dates=['Month'],index_col='Month')
ax=aq1.plot(figsize=(10,5),color='purple')
ax.set_ylabel('Value')
ax.set_xlabel('Year')
ax.set_title('BSE Index')
ax3=aq3.plot(figsize=(10,5),color='green')
ax3.set_ylabel('Price')
ax3.set_title('Crude Oil Price')
ax3.set_xlabel('Year')
ax2=aq2[['Mumbai Rupees per 10gms','London Rupees per 10gms.']].plot(figsize=(10,5))
ax2.set_ylabel('Value')
ax2.set_title('Gold Price in Mumbai & London')
ax2.set_xlabel('Year')
ax.plot()
ax3.plot()
ax2.plot()
plt.show()
```
